## Definição do Problema

O objetivo deste projeto é analisar dados de incêndios florestais para identificar padrões relacionados à ocorrência e intensidade dos incêndios.

Trata-se de um problema de análise exploratória (não supervisionado), cujo foco é compreender as relações entre variáveis climáticas e a área queimada.

**Hipóteses:**
- Temperatura influencia a propagação do fogo
- Baixa umidade aumenta o risco
- O vento contribui para a expansão


```
import pandas as pd

url = 'https://raw.githubusercontent.com/CarlysbergBarreto/mvp-analise-de-dados-incendios-florestais/main/dataset/forestfires.csv'
df = pd.read_csv(url)
df.head()
```

## Análise Exploratória

### Análise da Temperatura

Observa-se que a temperatura apresenta distribuição aproximadamente normal, indicando que a maioria dos incêndios ocorre em condições intermediárias.

Isso sugere que a temperatura isoladamente não é suficiente para explicar incêndios de grande escala.


## Análise por Estação do Ano

Observa-se que a maior incidência de incêndios ocorre durante o verão, período caracterizado por altas temperaturas e baixa umidade.

Durante o inverno, a ocorrência é significativamente menor devido à maior umidade.

A análise revelou que a maioria dos incêndios apresenta baixa área queimada, enquanto poucos eventos geram grande impacto.

Fatores climáticos como temperatura, umidade e vento influenciam significativamente a propagação do fogo.
