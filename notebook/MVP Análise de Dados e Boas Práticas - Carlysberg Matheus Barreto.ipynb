## Definição do Problema

Analisar dados de incêndios florestais para identificar padrões relacionados à ocorrência e intensidade.

Problema de análise exploratória (não supervisionado).


```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/CarlysbergBarreto/mvp-analise-de-dados-incendios-florestais/main/dataset/forestfires.csv'
df = pd.read_csv(url)
df.head()
```

## Análise Climática Cruzada

### Chuva vs Área Queimada

A precipitação atua como fator limitante na propagação do fogo.


```
sns.scatterplot(x='rain', y='area', data=df)
plt.show()
```

Observa-se que maiores níveis de chuva estão associados a menor área queimada.


### Umidade vs Área Queimada

A umidade influencia diretamente a propagação do fogo.


```
sns.scatterplot(x='RH', y='area', data=df)
plt.show()
```

Baixa umidade está associada a maiores áreas queimadas.


### Temperatura e Umidade combinadas


```
sns.scatterplot(x='temp', y='RH', hue='area', data=df)
plt.show()
```

## Conclusão

A análise mostra que incêndios estão associados a alta temperatura e baixa umidade.

A chuva reduz significativamente a propagação do fogo.
